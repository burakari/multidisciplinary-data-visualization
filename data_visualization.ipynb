# Multidisciplinary Data Visualization with Python

This project visualizes how engineering, law, and data science intersect through a WordCloud.

Author: Burak ARI

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
base_text = """
burak burak burak burak burak burak burak burak burak burak

data data data data data data data
python python python python
analytics analytics analytics
machine learning machine learning

engineering engineering engineering
mechanical aerospace systems thermodynamics dynamics

law legal reasoning compliance

sql statistics modeling regression classification clustering
automation optimization feature engineering data pipelines
pandas numpy matplotlib scikit-learn tensorflow keras

systems thinking interdisciplinary innovation artificial intelligence
"""

text = base_text * 50
```

```python
wc = WordCloud(
    width=2200,
    height=1100,
    background_color="#061F1F",
    stopwords=set(STOPWORDS),
    max_words=2000,
    margin=1,
    scale=4,
    prefer_horizontal=0.9,
    relative_scaling=0.22,
    min_font_size=6,
    collocations=False
).generate(text)
```

```python
def brand_colors(word, font_size, position, orientation, random_state=None, **kwargs):
    if word.lower() in ["burak", "data"]:
        return "#F97316"
    return "#FFFFFF"

wc.recolor(color_func=brand_colors)
```

```python
plt.figure(figsize=(22,11))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.tight_layout()

plt.savefig("output.png", dpi=300)
plt.show()
```
